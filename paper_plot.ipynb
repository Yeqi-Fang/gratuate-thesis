```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.ticker import MaxNLocator
```

```python
# Set the style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_context("notebook", font_scale=1.2)
```

```python
# Data
epochs = list(range(1, 31))
train_loss = [1.446493, 0.617663, 0.175501, 0.143573, 0.133605, 0.126618, 0.121166, 
              0.116923, 0.113489, 0.110262, 0.107590, 0.101929, 0.100653, 0.099758, 
              0.098903, 0.097319, 0.096991, 0.096707, 0.096465, 0.096016, 0.095908, 
              0.095830, 0.095770, 0.095578, 0.095550, 0.095544, 0.095529, 0.095472, 
              0.095443, 0.095451]
val_loss = [0.817413, 0.175731, 0.162331, 0.160131, 0.160416, 0.157523, 0.156998, 
            0.158202, 0.157653, 0.157364, 0.157302, 0.157555, 0.157721, 0.157249, 
            0.158251, 0.158667, 0.157865, 0.157359, 0.159177, 0.157912, 0.158210, 
            0.157368, 0.157229, 0.159537, 0.157774, 0.157889, 0.158049, 0.159883, 
            0.158214, 0.157624]

# Create figure and axis with a beautiful color palette
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the training and validation loss with beautiful colors
train_line, = ax.plot(epochs, train_loss, linewidth=3.0, marker='o', 
                     markersize=6, alpha=0.9, label='Training Loss', color='#3498db')
val_line, = ax.plot(epochs, val_loss, linewidth=3.0, marker='o', 
                   markersize=6, alpha=0.9, label='Validation Loss', color='#e74c3c')

# Set up the axes and labels
ax.set_xlabel('Epoch', fontsize=14, fontweight='bold')
ax.set_ylabel('MAE Loss (log scale)', fontsize=14, fontweight='bold')
ax.set_yscale('log')  # Using log scale for y-axis
ax.xaxis.set_major_locator(MaxNLocator(integer=True))  # Only use integer ticks for x-axis

# Add a title
plt.title('Training and Validation Loss', fontsize=16, fontweight='bold', pad=20)

# Add legend
ax.legend(fontsize=12)

# Add a grid (with more visibility on log scale)
ax.grid(True, alpha=0.3, which='both')  # 'both' ensures grid lines on major and minor ticks

# Adjust layout and show the plot
plt.tight_layout()
plt.savefig('paper_image/loss_plot.pdf')
plt.show()
```
